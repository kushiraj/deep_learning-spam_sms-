```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Conv1D, LSTM, GRU, Bidirectional, Dense, Flatten, Dropout, GlobalMaxPooling1D
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Load dataset
df = pd.read_csv('/content/spam_sms.csv')
df.columns = ['label', 'message']  # Assuming two columns: label (spam/ham) and message

# Convert labels to binary (1 for spam, 0 for ham)
df['label'] = df['label'].map({'spam': 1, 'ham': 0})

# Text preprocessing
MAX_WORDS = 5000  # Vocabulary size
MAX_LEN = 100  # Maximum sequence length
tokenizer = Tokenizer(num_words=MAX_WORDS, oov_token="<OOV>")
tokenizer.fit_on_texts(df['message'])
sequences = tokenizer.texts_to_sequences(df['message'])
padded_sequences = pad_sequences(sequences, maxlen=MAX_LEN, padding='post')

# Splitting data
X_train, X_test, y_train, y_test = train_test_split(padded_sequences, df['label'], test_size=0.2, random_state=42)

# Function to build models
def build_model(model_type="CNN"):
    model = Sequential()
    model.add(Embedding(MAX_WORDS, 128, input_length=MAX_LEN))
    if model_type == "CNN":
        model.add(Conv1D(64, 5, activation='relu'))
        model.add(GlobalMaxPooling1D())
    elif model_type == "LSTM":
        model.add(LSTM(64, return_sequences=False))
    elif model_type == "GRU":
        model.add(GRU(64, return_sequences=False))
    elif model_type == "BiLSTM":
        model.add(Bidirectional(LSTM(64, return_sequences=False)))
    elif model_type == "BERT":
        # Placeholder for BERT model (Would need transformers library)
        return None
    elif model_type == "CNN-LSTM-LSTM":
        model.add(Conv1D(64, 5, activation='relu'))
        model.add(LSTM(64, return_sequences=True))
        model.add(LSTM(64, return_sequences=False))
    model.add(Dense(64, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=0.001), metrics=['accuracy'])
    return model

# Train and evaluate models
models = ["CNN", "LSTM", "GRU", "BiLSTM", "CNN-LSTM-LSTM"]
results = {}
history_dict = {}

for model_type in models:
    print(f"\nTraining {model_type} model...")
    model = build_model(model_type)
    history = model.fit(X_train, y_train, epochs=5, batch_size=32, validation_data=(X_test, y_test), verbose=1)
    y_pred = (model.predict(X_test) > 0.5).astype("int32")
    results[model_type] = {
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred),
        "Recall": recall_score(y_test, y_pred),
        "F1-Score": f1_score(y_test, y_pred)
    }
    history_dict[model_type] = history

    print(f"\nResults for {model_type} model:")
    for metric, value in results[model_type].items():
        print(f"{metric}: {value:.4f}")

# Plot training accuracy and loss for all models
plt.figure(figsize=(12, 6))
for model_type, history in history_dict.items():
    plt.plot(history.history['accuracy'], label=f'{model_type} Train Accuracy')
    plt.plot(history.history['val_accuracy'], label=f'{model_type} Val Accuracy')
plt.title('Model Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

plt.figure(figsize=(12, 6))
for model_type, history in history_dict.items():
    plt.plot(history.history['loss'], label=f'{model_type} Train Loss')
    plt.plot(history.history['val_loss'], label=f'{model_type} Val Loss')
plt.title('Model Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
